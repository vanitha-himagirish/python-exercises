Checking whether phone numbers of a given employee are valid - get_invalid_phone_count
Function should take 2 arguments, employee_id and phone_numbers (variable number)
Check whether each phone number have 10 digits.
Return employee_id and number of phone numbers with less than 10 digits


```python
def get_invalid_phone_count(employee_id,*phone_numbers):
    invalid=0;
    for ph in phone_numbers:
        if len(ph)!=10:
            invalid=invalid+1
    return employee_id,invalid
            
```

```python
get_invalid_phone_count(123,'1234','2345678901','12345')
```

```python
s = 'Employee {employee_id} have {invalid_count} invalid phones'
```

```python
employee_id, invalid_count = get_invalid_phone_count(1, '1234567890', '1234','234')
print(s.format(employee_id=employee_id, invalid_count=invalid_count))

```

function code with out lambda function - Sum of squares of integers between a range


```python
def sumOfSquares(ub,lb):
    total=0
    for i in range(lb,ub+1):
        total+=i*i
    return total

sumOfSquares(3,1)
        
```

Function code with lambda function - Sum of squares of integers between a range

```python
def my_sum(lb, ub, f):
    total = 0
    for i in range(lb, ub + 1):
        total += f(i)
    return total

my_sum(1, 3, lambda n: n * n)
```

```python
my_sum(1, 3, lambda n: n * n * n)
```
