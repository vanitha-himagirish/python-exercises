**Get number of COMPLETE orders placed by each customer


```python
def get_orders_completed(customer_id,orders):
    total_orders=0
    for order in orders:
        if int(order.split(',')[2]) == customer_id and (order.split(',')[3] == 'COMPLETE' or order.split(',')[3] == 'CLOSED'):
            total_orders=total_orders+1
    return total_orders;

get_orders_completed(12111,orders)
            
            
            
```

Get total number of PENDING or PENDING_PAYMENT orders for the month of 2014 January.


```python
def get_orders_pending(customer_id,orders,order_month):
    total_orders=0
    for order in orders:
        if int(order.split(',')[2]) == customer_id and (order.split(',')[3] == 'PENDING' or order.split(',')[3] == 'PENDING_PAYMENT') and order.split(',')[1].startswith(order_month):
            total_orders=total_orders+1
    return total_orders;

get_orders_pending(10778,orders,'2014-01')
```

Get outstanding amount for each month considering orders with status PAYMENT_REVIEW, PENDING, PENDING_PAYMENT and PROCESSING.

```python
path = '/home/vanithah/data/order_items.csv' 
orders_file = open(path)
orders_raw = orders_file.read()
order_items = orders_raw.splitlines()
```

```python
def get_orders_pending(orders,*order_status):
    order_month={}
    for order in orders:
         if (order.split(',')[3] in order_status):
                order_count=int(order.split(',')[0])            
                order_month[order_count] = order.split(',')[1]
    return order_month;

get_orders_pending(orders,'PAYMENT_REVIEW','PENDING','PENDING_PAYMENT','PROCESSING')
```

```python
def get_outstanding_amount(orders, order_items, *order_status):
    orders_dict = get_orders_pending(orders,*order_status)
    outstanding_amount = {}
    for order_item in order_items:
        order_item_order_id = int(order_item.split(',')[1])
        order_item_subtotal = float(order_item.split(',')[4])
        if order_item_order_id in orders_dict:
            orders_dict_date = orders_dict[order_item_order_id][:7]
            
            if orders_dict_date in outstanding_amount:
                outstanding_amount[orders_dict_date] = round(outstanding_amount[orders_dict_date] + order_item_subtotal, 2)
            else:
                outstanding_amount[orders_dict_date] = order_item_subtotal
    return outstanding_amount
```

```python
get_outstanding_amount(orders, order_items, 'PAYMENT_REVIEW','PENDING','PENDING_PAYMENT','PROCESSING')
```
